```python
import numpy as np
import matplotlib.pyplot as plt
from complex_iteration import mandelbrot_iteration
from matplotlib.colors import ListedColormap, BoundaryNorm
```

# Question 1

```python
#Create a meshgrid to define c
x = np.linspace(-2, 2, 1000)
y = x
x, y = np.meshgrid(x, y)
i = 1j
c = x + y * i #define each c on cartesian plane
```

```python
max_iter = 1000 #maximum number of iterations
arb_limit = 1e6 #arbitrary limit for |z|
z = 0
passed = np.zeros(c.shape)
```

```python
z, conv = mandelbrot_iteration(c) #perform the mandelbrot iteration
```

```python
plt.figure(dpi = 150)
plt.contourf(x, y, abs(z))
plt.axis('scaled')
plt.colorbar(label = '$|z|$')

plt.xlabel('$c_x$')
plt.ylabel('$c_y$')
plt.suptitle(rf'$|z|$ for a Maximum of {max_iter} Iterations', fontweight = 'bold')
plt.title('(divergent regions in white)', fontsize = 9)
plt.show()
```

```python
# Define a two-color colormap (e.g., red and blue)
cmap = ListedColormap(["#EE6C4D", "#3D5A80"])

# Define boundaries for 0 and 1
bounds = [-.5, 0.5, 1.5]
norm = BoundaryNorm(bounds, cmap.N)
```

```python
valid_mask = ~(np.isnan(z.real) | np.isnan(z.imag))
c_conv = c[valid_mask]
```

```python
plt.figure(dpi = 150)
plt.contourf(x, y, conv, cmap = cmap, norm = norm)
plt.axis('scaled')
cbar = plt.colorbar()
cbar.set_ticks([0, 1])          # Set tick positions
cbar.set_ticklabels(['Divergent', 'Convergent']) #set tick labels

plt.xlabel('$c_x$')
plt.ylabel('$c_y$')
plt.suptitle(r'Convergent and Divergent Regions for $z_{{n+1} = z_n + c$', fontweight = 'bold')
plt.show()
```
